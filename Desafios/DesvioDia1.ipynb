# Desafio 6 - Renomear as colunas tirando o hifen

## solução 1

```python
dados_temp = dados.copy() # eu nao quero modificar o dataFrame original

new_columns = []
for word in dados_temp.columns:
  if "-" in word:
    new_columns.append(word.replace("-", ""))
  else:
    new_columns.append(word)

dados_temp.columns = new_columns

dados_temp.head()
```

## solução 2

# Desafio 7 - Deixar os gráficos bonitos

```python
import matplotlib.pyplot as plt

labels = 'Com droga', 'Com controle'
ax = dados['tratamento'].value_counts().plot.pie(shadow=False,                                                
                                                labels=labels,  
                                                figsize=(10, 8),
                                                explode=[0,0.2],
                                                autopct='%1.1f%%' )

ax.set_title("Proporção do tipo de tratamento", fontsize=24)
ax.set_ylabel("")
plt.legend()
plt.show()
```

```python
ax= dados['tempo'].value_counts().plot.bar(figsize=(10, 8))
ax.set_title("Número de testes por tempo de dosagem", fontsize=24)
ax.set_ylabel("Quantidade")
ax.set_xlabel("Horas")
plt.show()
```
